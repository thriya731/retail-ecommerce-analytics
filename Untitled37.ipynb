```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
orders=pd.read_csv("/content/orders (1).csv")
products=pd.read_csv("/content/products (1).csv")
customers=pd.read_csv("/content/customers (1).csv")
```

step 1: Data Cleaning

```python
orders.info()
orders.describe()
orders.isnull().sum()
orders.duplicated().sum()
orders['ship_date'] = pd.to_datetime(orders['ship_date'])
orders['ship_date'] = orders['ship_date'].fillna(orders['ship_date'].median())
orders['sales_rep_id'] = orders['sales_rep_id'].fillna(0)
orders['revenue_usd'] = orders['quantity'] * orders['unit_price_usd']
orders['order_status'] = orders['order_status'].str.capitalize()
orders['payment_method']=orders['payment_method'].str.capitalize()
orders['ship_date_error']=(orders['ship_date']==0)

products.info()
products.describe()
products.isnull().sum()
products['cost_usd'] = products['cost_usd'].fillna(products['cost_usd'].median())

customers.info()
customers['first_name']=customers['first_name'].str.capitalize()
customers['email']=customers['email'].fillna(0)
customers['lifetime_value_usd']=customers['lifetime_value_usd'].fillna(customers['lifetime_value_usd'].median())
customers['region']=customers['region'].str.capitalize()
```

step 2: EDA

```python
a=pd.merge(orders,customers,on='customer_id',how='left')
final_table=pd.merge(a,products,on='product_id',how='left')
final_table['order_date']=pd.to_datetime(final_table['order_date'], format='mixed', dayfirst=True)
final_table['year-month']=final_table['order_date'].dt.strftime('%Y-%m')
final_table['revenue_usd']=(final_table['quantity']*final_table['unit_price_usd_x'])*(100-final_table['discount_pct'])/100
final_table['discount_amt_usd']=(final_table['quantity']*final_table['unit_price_usd_x'])*5/100
```

```python
churn_table['latest_order_date']=pd.to_datetime(churn_table['latest_order_date'])
churn_table['gap'] = churn_table['gap'].dt.days
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.utils import resample
from sklearn.model_selection import GridSearchCV
```

```python
churn_table
```

```python
churn_table['latest_order_date']= (
    churn_table['latest_order_date'].astype(int) / 10**9)
x=churn_table.drop(columns=['flag','Churn'])
y=churn_table['Churn']
x=pd.get_dummies(x,drop_first=True)
X_train,X_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.2)

train = pd.concat([X_train, y_train], axis=1)
train['Churn'].value_counts()
major=train[train['Churn']==0]
minor=train[train['Churn']==1]
minor_unsampled=resample(minor,n_samples=len(major),random_state=42,replace=True)
final=pd.concat([major,minor_unsampled],axis=0)
final_x=final.drop(columns=['Churn'])
final_y=final['Churn']
model=RandomForestClassifier(random_state=42,class_weight='balanced')
params={'n_estimators':[100,150]}
grid=GridSearchCV(model,params,cv=5)
grid.fit(final_x,final_y)
y_pred=grid.predict(X_test)
proba=grid.predict_proba(X_test)[:,1]
print(proba[:10])
threshold=0.8
(proba>threshold).astype(int)
print('accuracy',accuracy_score(y_test,y_pred))
print('classification',classification_report(y_test,y_pred))
print('confusion',confusion_matrix(y_test,y_pred))
```

```python
best_model = grid.best_estimator_

importance = pd.DataFrame({
    'feature': final_x.columns,
    'importance': best_model.feature_importances_
})

print(importance.sort_values(by='importance', ascending=False))
```
